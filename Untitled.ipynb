```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([('ohe',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','venue'])],remainder='passthrough')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=match.iloc[:,:9]
```

```python
y=match.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear',max_iter=5000))])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
def match_progression(df,id,pipe):
    df=df[df['id']==id]
    df=df[df['balls_left']%6==0]
    result=pipe.predict_proba(df)
    df['end_of_over']=20-(df['balls_left']/6)
    df['runs_after_over']= df['score'] - df['score'].shift(1, fill_value=0)
    df['wickets_in_over']=df['wickets_left']-df['wickets_left'].shift(-1,fill_value=0)
    
    df['loose']=np.round(result.T[0]*100,1)
    df['win']=np.round(result.T[1]*100,1)
    target=df['target_runs'].values[0]
    print(target)
    df=df[['end_of_over','runs_after_over','wickets_in_over','score','loose','win']]
    return df
    
    
    
```

```python
temp_df=match_progression(match_pro,829795,pipe)
```
